```python
# What is the survival rate based on Pclass?

survived = titanic[titanic['Survived'] == 1]
died = titanic[titanic['Survived'] == 0]
```

```python
survived
```

```python
# Count how many times each unique vaue in a column occurrs
survived['Pclass'].value_counts()
```

```python
died['Pclass'].value_counts()
```

```python
plt.figure(figsize=(4,3))
survived['Pclass'].value_counts().plot(kind='bar')
plt.title('Survival Rate by PClass\n')
plt.ylabel('Survival Count')

plt.savefig('survival_by_pclass.png')
```

```python
died['Pclass'].value_counts().plot(kind='bar')
```

```python
results = pd.concat([survived['Pclass'].value_counts(), died['Pclass'].value_counts()], axis='columns')
results
```

```python
from sklearn.preprocessing import (
    LabelEncoder, OneHotEncoder, 
    StandardScaler, MinMaxScaler)

from pandas import get_dummies
```

```python
titanic.head()
```

```python
columns = ['Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Survived']
data = titanic[columns]
```

```python
data.head()
```

```python
data.info()
```

```python
# Counting missing values
data['Age'].isna().sum()
```

```python
data[data['Age'].isna()]
```

**How to address missing values**
- Remove rows with missing values
- Substitute mean value for missing numeric column
- Substitute median value for missing numeric column
- Use mode to sustitute for missing categorical column

```python
data['Age'].plot(kind='hist')
```

```python
# Filling-in the missing data for Age using the mean, since it is normally distributed with smal range
avg_age = data['Age'].mean(skipna=True)

data['Age'].fillna(avg_age, inplace=True)
```
